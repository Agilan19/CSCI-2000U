```python
#Agilan Ampigaipathar
#100553054

#Plot inline
%pylab inline

#Imports
from struct import *
import math as math
import matplotlib.pyplot as plt
import numpy as np

#Make a 10 x 10 figure to plot on
plt.figure(figsize=(10, 10))

#Open the instructed file
with open('/home/agilan/csci-2000-personal/Large_Data/halo.dat','rb') as file1:
    N_halo = file1.readline()
    T_halo = file1.readline()
    data_halo = file1.read()
    
    unpacked_N = unpack("i"*35, N_halo[:-2])
    unpacked_T = unpack("d"*31, T_halo[:-3])
    unpacked_data = unpack("i" * ((len(data_halo) -3) // 4), data_halo[:-3])
    vx = []
    vy = []
    vz = []
    v_halo = []
    count = len(unpacked_data)
    
    i = 0
    for j in unpacked_data:
        
        #I manually took the count value and divided it by 4 and rounded
        #down to get 401661.
        if(i == 401661):
            break
            
        elif(i%3 == 0):
            vx.append(j)
            i = i + 1
            
        elif(i%3 == 1):
            vy.append(j)
            i = i + 1
            
        elif(i%3 == 2):
            vz.append(j)
            i = i + 1
            
    k = 0    
    
    while k < (i/3):
        v_halo.append(math.sqrt((vx[k]**2) + (vy[k]**2) + (vz[k]**2)))
        k = k + 1

hist, bins = np.histogram(v_halo, bins = 50)
width = 0.7 * (bins[1] - bins[0])
center = (bins[:-1] + bins[1:]) / 2
plt.bar(center, hist, width=width, color= 'green')
plt.ylabel('Velocity')
plt.title('Velocity histogram (halo) (100553054)')
plt.show()

#Make a 10 x 10 figure to plot on
plt.figure(figsize=(10, 10))

#Open the instructed file
with open('/home/agilan/csci-2000-personal/Large_Data/snapshot_010.dat', 'rb') as file2:
    N_010 = file2.readline()
    T_010 = file2.readline()
    data_010 = file2.read()
    
    count_N = len(N_010)
    count_T = len(T_010)
    count_data = len(data_010)
    
    unpacked_N = unpack("i"*(math.floor(count_N/4)), N_010[:(-1*(count_N%4))])
    unpacked_T = unpack("d"*(math.floor(count_T/8)), T_010[:(-1*(count_T%8))])
    unpacked_data = unpack("i" * (math.floor((count_data)/ 4)), data_010[:(-1*(count_data%4))])
    vx = []
    vy = []
    vz = []
    v_010 = []
    
    i = 0
    for j in unpacked_data:
        
        #I manually took the count_data value and divided it by 4 and 
        #to get 14679642
        if(i == 14679642):
            break
            
        elif(i%3 == 0):
            vx.append(j)
            i = i + 1
            
        elif(i%3 == 1):
            vy.append(j)
            i = i + 1
            
        elif(i%3 == 2):
            vz.append(j)
            i = i + 1
            
    k = 0    
    
    while k < (i/3):
        v_010.append(math.sqrt((vx[k]**2) + (vy[k]**2) + (vz[k]**2)))
        k = k + 1

hist, bins = np.histogram(v_010, bins = 50)
width = 0.7 * (bins[1] - bins[0])
center = (bins[:-1] + bins[1:]) / 2
plt.bar(center, hist, width=width, color= 'blue')
plt.ylabel('Velocity')
plt.title('Velocity histogram (010) (100553054)')
plt.show()
```
