```python
#Agilan Ampigaipathar
#100553054

#Set numpy as np so np could be used instead later in the program.
import numpy as np

#Import the matplotlib.pyplot package as plt to be later used as plt 
#instead.
import matplotlib.pyplot as plt

#Will cause the plot to appear as inline output.
%matplotlib inline

#Make a large figure box to hold the plot. The larger the figure size,
#the larger the plot and it becomes more readable and clear.
plt.figure(figsize=(10, 10))

#Load the data within the .dat file into the data variable,
data = np.loadtxt('xy_points.dat')

#Saves x and y values from the data into variables.
x = data[0][:]
y = data[1][:]

#Acquires the values for x and y when x is less than -2
x1 = x[x < -2]
y1 = y[x < -2]

#Gathers the values for x and y when the updated y is is greater 
#or equal to -5
x2 = x1[-5 <= y1]
y2 = y1[-5 <= y1]

#Gets the values for x and y when the updated y is less than or equal 
#to 0.
x3 = x2[y2 <= 0]
y3 = y2[y2 <= 0]

#Make xc and yc values based on the coordinate assigned. 
xc = 0
yc = 5

#Calculate for delta x and y
deltax = x-xc
deltay = y-yc

#Use the formula given to calculate for r.
r = np.sqrt(((deltax)**2)+((deltay)**2))

#Gathers only the x and y values that are less than 5.
x_new = x[r < 5]
y_new = y[r < 5]

#Set limits from -10 to 10 for x and y as asked in the assignment. 
plt.ylim(-10,10)
plt.xlim(-10,10) 

#Plot the red dots by plotting the x of the data against the y.
plt.scatter(x, y, c = 'red')

#Plot the blue dots by plotting the updated x against the updated y
#after passing through the parameters. 
plt.scatter(x3, y3, c = 'blue')

#Plot the green dots by plotting the new x against the new y values 
#that makes a green circle as described in the assignment doc. 
plt.scatter(x_new,y_new, c = 'green')

#Make the assigned title.
plt.title('Highlighted points (100553054)')

#Outputs the results of the final plot. 
plt.show()
```
